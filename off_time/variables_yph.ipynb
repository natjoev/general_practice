```python
# import libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

# set file path 
path = "/Users/ael/OneDrive/Desktop/138/03_data/capstone"
```

```python
vol_crop = pd.read_excel(
    path + "/palay.xlsx", 
    skiprows=[1, 2, 3, 4], 
    usecols="B:L", # to clean up the data
    header=0
)
vol_crop.dropna(inplace=True)

vol_crop 
```

```python
# rename header
vol_crop.rename(
    columns={
    "Unnamed: 1": "Region",
    "Unnamed: 2": "2015v_vop",
    'Unnamed: 3': "2016_vop",
    'Unnamed: 4': "2017_vop",
    'Unnamed: 5': "2018_vop",
    'Unnamed: 6': "2019_vop",
    'Unnamed: 7': "2020_vop",
    'Unnamed: 8': "2021_vop",
    'Unnamed: 9': "2022_vop",
    'Unnamed: 10': "2023_vop",
    'Unnamed: 11': "2024_vop",
}, inplace=True
)
vol_crop
```

```python
# Do same for area data set
# since data structure is the same, we can just copy the previous code and modify it
area_crop = pd.read_excel(
    path + "/palay_area.xlsx", 
    skiprows=[1, 2, 3, 4], 
    usecols="B:L", # to clean up the data
    header=0
)
area_crop.rename(
    columns={
    "Unnamed: 1": "Region",
    "Unnamed: 2": "2015v_aih",
    'Unnamed: 3': "2016_aih",
    'Unnamed: 4': "2017_aih",
    'Unnamed: 5': "2018_aih",
    'Unnamed: 6': "2019_aih",
    'Unnamed: 7': "2020_aih",
    'Unnamed: 8': "2021_aih",
    'Unnamed: 9': "2022_aih",
    'Unnamed: 10': "2023_aih",
    'Unnamed: 11': "2024_aih",
}, inplace=True
)
area_crop.dropna(inplace=True)
area_crop
```

```python
# concat by region
palay_data = pd.concat([vol_crop, area_crop.iloc[:, 1:]], axis=1)
palay_data.dropna(inplace=True)
palay_data
```

```python
# make new df for yield per hectare
# first make varaiabls for yield per hectare for years 2015-2024
# make sure that data is not string 
palay_data["2015v_vop"] = pd.to_numeric(palay_data["2015v_vop"], errors='coerce')
palay_data["2016_vop"] = pd.to_numeric(palay_data["2016_vop"], errors='coerce')
palay_data["2017_vop"] = pd.to_numeric(palay_data["2017_vop"], errors='coerce')
palay_data["2018_vop"] = pd.to_numeric(palay_data["2018_vop"], errors='coerce')
palay_data["2019_vop"] = pd.to_numeric(palay_data["2019_vop"], errors='coerce')
palay_data["2020_vop"] = pd.to_numeric(palay_data["2020_vop"], errors='coerce')
palay_data["2021_vop"] = pd.to_numeric(palay_data["2021_vop"], errors='coerce')
palay_data["2022_vop"] = pd.to_numeric(palay_data["2022_vop"], errors='coerce')
palay_data["2023_vop"] = pd.to_numeric(palay_data["2023_vop"], errors='coerce')
palay_data["2024_vop"] = pd.to_numeric(palay_data["2024_vop"], errors='coerce')
palay_data["2015v_aih"] = pd.to_numeric(palay_data["2015v_aih"], errors='coerce')
palay_data["2016_aih"] = pd.to_numeric(palay_data["2016_aih"], errors='coerce')
palay_data["2017_aih"] = pd.to_numeric(palay_data["2017_aih"], errors='coerce')
palay_data["2018_aih"] = pd.to_numeric(palay_data["2018_aih"], errors='coerce')
palay_data["2019_aih"] = pd.to_numeric(palay_data["2019_aih"], errors='coerce')
palay_data["2020_aih"] = pd.to_numeric(palay_data["2020_aih"], errors='coerce')
palay_data["2021_aih"] = pd.to_numeric(palay_data["2021_aih"], errors='coerce')
palay_data["2022_aih"] = pd.to_numeric(palay_data["2022_aih"], errors='coerce')
palay_data["2023_aih"] = pd.to_numeric(palay_data["2023_aih"], errors='coerce')
palay_data["2024_aih"] = pd.to_numeric(palay_data["2024_aih"], errors='coerce')

# make variables for yield per hectare
palay_data ["2015_yph"] = palay_data["2015v_vop"] / palay_data["2015v_aih"]
palay_data ["2016_yph"] = palay_data["2016_vop"] / palay_data["2016_aih"]
palay_data ["2017_yph"] = palay_data["2017_vop"] / palay_data["2017_aih"]
palay_data ["2018_yph"] = palay_data["2018_vop"] / palay_data["2018_aih"]
palay_data ["2019_yph"] = palay_data["2019_vop"] / palay_data["2019_aih"]
palay_data ["2020_yph"] = palay_data["2020_vop"] / palay_data["2020_aih"]
palay_data ["2021_yph"] = palay_data["2021_vop"] / palay_data["2021_aih"]
palay_data ["2022_yph"] = palay_data["2022_vop"] / palay_data["2022_aih"]
palay_data ["2023_yph"] = palay_data["2023_vop"] / palay_data["2023_aih"]
palay_data ["2024_yph"] = palay_data["2024_vop"] / palay_data["2024_aih"]
palay_data
```
